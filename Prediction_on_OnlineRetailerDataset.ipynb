```python
import numpy as np
import pandas as pd
import os
# path of the folder that contains the datasets
folderpath = r'C:\Users\egenur' 
filename = 'OnlineRetailerDataset.csv'
#filepath = folderpath + '\\' + filenames
filepath = os.path.join(folderpath, filename)
filepath
```

```python
dfglass = pd.read_csv(filepath,index_col='User_ID') 
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()

le.fit(dfglass['Gender'])
dfglass['Gender'] = le.transform(dfglass['Gender'])

le.fit(dfglass['Marital_Status'])
dfglass['Marital_Status'] = le.transform(dfglass['Marital_Status'])

               
le.fit(dfglass['Browsed_Electronics_12Mo'])
dfglass['Browsed_Electronics_12Mo'] = le.transform(dfglass['Browsed_Electronics_12Mo'])
               
le.fit(dfglass['Bought_Electronics_12Mo'])
dfglass['Bought_Electronics_12Mo'] = le.transform(dfglass['Bought_Electronics_12Mo'])
               
               
le.fit(dfglass['Bought_Digital_Media_18Mo'])
dfglass['Bought_Digital_Media_18Mo'] = le.transform(dfglass['Bought_Digital_Media_18Mo'])
               
            
le.fit(dfglass['Bought_Digital_Books'])
dfglass['Bought_Digital_Books'] = le.transform(dfglass['Bought_Digital_Books'])
               
le.fit(dfglass['eReader_Adoption'])
dfglass['eReader_Adoption'] = le.transform(dfglass['eReader_Adoption'])
               
                
le.fit(dfglass['Payment_Method'])
dfglass['Payment_Method'] = le.transform(dfglass['Payment_Method'])
          
le.fit(dfglass['Website_Activity'])
dfglass['Website_Activity'] = le.transform(dfglass['Website_Activity'])
```

```python
dfglass['Website_Activity'].head()
```

```python
X = dfglass.drop(['eReader_Adoption'], axis=1) # axis is 1 since we drop a column
y = dfglass['eReader_Adoption']
```

```python
X.head()
```

```python

from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier    #DESICIONTREECLASSIFIER
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB           
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score #prediction calculations

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
def BoldPrint(string):
    print(color.BOLD + string + color.END)
```

```python
#Kfold cross validation
k = 10
skf = KFold(n_splits=k)
accuracy_results = []
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    clf = DecisionTreeClassifier()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)    
    accuracy_results.append(acc)
print("Accuracy results of each iteration:", accuracy_results)
print("Aggregated accuracy result:", np.average(accuracy_results))
```

```python
classifiers = [
    KNeighborsClassifier(3),
    DecisionTreeClassifier(max_depth=3),
    SVC(gamma=2, C=1),
    RandomForestClassifier( n_estimators=100),
    MLPClassifier( max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
   QuadraticDiscriminantAnalysis()
]
classifierNames = [
    "KneighborsClassifier",
    "DecisionTreeClassifier(max_depth=3)",
    "SVC( gamma=2 , C=1 )",
    "RandomForestClassifier( n_estimators=100 )",
    "MLPClassifier(  max_iter=1000 )",
    "AdaBoostClassifier()",
    "GaussianNB()",
   "QuadraticDiscriminantAnalysis()"
    
]
classifierRange=len(classifierNames)
print("Number of classifiers that will be tested:", classifierRange)
```

```python
# Classifier Metric Calculations
from sklearn.model_selection import StratifiedKFold, KFold
import math
k = 10
skf = StratifiedKFold(n_splits=k,random_state=0)
classifierAccuracyDic=  dict()
for i in range(0,classifierRange):
    accuracy_results = []
    recall_results = []
    precision_results = []
    f1_results = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf = classifiers[i]
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        # should consider other performance metrics
        acc = accuracy_score(y_test, y_pred)    
        accuracy_results.append(acc)
        
        rec=recall_score(y_test, y_pred,average=None)    
        recall_results.append(rec)
        pr= precision_score(y_test,y_pred,average="weighted",labels=np.unique(y_pred))
        precision_results.append(pr)
        f1=f1_score(y_test,y_pred,average="weighted",labels=np.unique(y_pred))
        f1_results.append(f1)    
    BoldPrint("For "+classifierNames[i])
    print('{:<44} '.format( "Accuracy results of iterations" ),":",accuracy_results)
    print('{:<45} '.format( "\nAggregated accuracy result" ),":",end="")
    print("{0:.2f}%".format(np.average(accuracy_results)*100)) 
    print('{:<45} '.format( "\nRecall Score " ),":",end="") 
    print("{0:.2f}%".format(np.average(recall_results)*100)) 
    print('{:<45} '.format( "\nPrecision Score" ),":",end="") 
    print("{0:.2f}%".format(np.average(precision_results)*100)) 
    print('{:<45} '.format( "\nf1 Score" ),":",end="") 
    print("{0:.2f}%".format(np.average(f1_results)*100)) 
    
    classifierAccuracyDic[ classifierNames[i] ] = np.average(accuracy_results)
    i+=1
```

```python
#Top 3 Classifiers with Aggregated Accuracy Results
from collections import Counter
k = Counter(classifierAccuracyDic) 
   
high = k.most_common(3)  

BoldPrint('Top 3 classifiers with the higest accuracy\n')
  
for i in high: 
    #print(i[0]," :",i[1]," ") 
    print('{:<45} '.format( str(i[0]) ),' :  %s'%( i[1] )) 
```
